```python
from p8 import *
```

```python
par = {
    'std1' : 3,
    'std2' : 5,
    'filter_size' : 13,
    'min_val' : 0,
    'max_val' : 255,
    'time' : 50,
    'dt' : 1  
}

img_dim = 60

images = []
for i in range(1, 21):
    images.append(Image.open(f'./images/{i}.jpg'))

images_tensors = [torch.tensor(np.asarray(
    img.convert('RGB').convert('L').resize((img_dim,img_dim)))) for img in images]

coded_imgs = [TimeToFirstSpikeEncoder(
    img_tensor, par['time'], par['dt'], par['min_val'],
    par['max_val']) for img_tensor in images_tensors]

fig = plt.figure(figsize=(10, 5))
rows = 4
columns = 5

for idx, img in enumerate(images_tensors):
    fig.add_subplot(rows, columns, idx + 1)
    plt.imshow(img, cmap='gray')
```

# 1

```python
params['lr'] = 0.05
params['feature_number'] = 4
params['feature_size'] = 10
features = torch.rand(params['feature_number'], params['feature_size'], params['feature_size']) * 0.1 + 0.1
train_(coded_imgs, features)
```
